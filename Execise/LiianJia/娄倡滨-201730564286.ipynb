```python
#以下求最大值，最小值，平均值均为Size,Price,Year的
data=[]
with open('C://Users//30141//Desktop//应用数据科学//lianjia//lianjia.csv',encoding='utf-8') as csvfile:
    for line in csvfile.readlines():
        if line.strip() != '':
            data.append(line.strip().split(','))
head,data=data[0],data[1:]
list_room,list_price,list_time=[],[],[]
for j in data:
    list_room.append(float(j[head.index('Size')])) 
    list_price.append(float(j[head.index('Price')])) 
    list_time.append(float(j[head.index('Year')]))
```

```python
def maxminave(list_):
    maxnum=max(list_)
    minnum=min(list_)
    average=sum(list_)/len(list_)
    return('最大值:',maxnum,'最小值:',minnum,'平均值:',average)
print('Size:',maxminave(list_room))
print('Time:',maxminave(list_time))
print('Price:',maxminave(list_price))
```

```python
def varandstd(list_):
    sum1=0
    average=sum(list_)/len(list_)
    for i in list_room:
        sum1+=(i-average)**2
    var=sum1/len(list_room)
    std=var**0.5
    return('方差:',var,'标准差:',std)
print('Size:',varandstd(list_room))
print('Time:',varandstd(list_time))
print('Price:',varandstd(list_price))
```

```python
def modenum(listroom):
    Dic = {}
    mode=0
    for i in list_room:
        if i in Dic:Dic[i] += 1
        else:Dic[i] = 1
    nums=max(Dic.values())
    mode=list(Dic.keys())[list(Dic.values()).index(nums)]
    return('众数:',mode)
print('Size:',modenum(list_room))
```

```python
def skekur(listroom):
    sum1,sum2,sum3=0,0,0
    average=sum(listroom)/len(listroom)
    for i in list_room:
        sum1+=(i-average)**2
    var=sum1/len(list_room)
    std=var**0.5
    for i in list_room:
        sum2+=((i-average)/std)**3
        skewness=sum2/len(list_room)
    for i in list_room:
        sum3+=((i-average)/std)**4
        kurtosis=sum3/len(list_room)
    return('偏度:',skewness,'峰度:',kurtosis)
print('Size:',skekur(list_room))
print('Time:',skekur(list_time))
print('Price:',skekur(list_price))
```

```python
def quar(listroom):
    list2=sorted(list_room)
    len1=len(list_room)
    Quartile1=list2[round(0.25*len1-1)]
    Quartile2=list2[round(0.5*len1-1)]
    Quartile3=list2[round(0.75*len1-1)]
    return('四分位数:',Quartile1,Quartile2,Quartile3,'中位数:',Quartile2)
print('Size:',quar(list_room))
```

```python
#单位房价与建成时间有关系么？我们可以做一个回归模型来验证，数学基础是最小二乘法OLS
time=[]
for j in data:
    time.append(2019-int(j[head.index('Year')]))
```

```python
#建立一个函数，用最小二乘法求a,b,R^2
def f(list1, list2):
    cov,var,ssr,sst=0,0,0,0
    avex,avey=sum(list1)/len(list1),sum(list2)/len(list2)
    for i in range(0,len(list1)):
        cov+=(list1[i]-avex)*(list2[i]-avey)
        var+=(list1[i]-avex)**2
    a,b=cov/var,avey-(cov/var)*avex
    for i in range(0,len(list1)):sst+=(list2[i]-avey)**2
    for i in list1:ssr+=(a*i+b-avey)**2
    a,b,r=round(a,4),round(b,4),round(sst/ssr,4)
    print ("y = {} * x + {}     R^2: {} ".format(a,b,r))
f(perprice,time)
#得出结果比较惊人。。越新的二手房，单位面积价格越低，但是R^2拟合结果也较差，说明残差中包含了很多可以解释因变量变化的因素。
#但是基于已有的信息，其他因素大都是非数值型，有点难以量化。
```
