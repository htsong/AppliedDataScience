```python
room=[]
floor_lst=[]
id_lst=[]
price_lst=[]
size_lst=[]
year_lst=[]
district_lst=[]
with open('lianjia.csv',encoding='utf8') as file:
    for line in file.readlines():
        if line.strip() is not '':room.append(line.strip().split(','))
roomhd,roomda=room[0],room[1:]
print(roomhd,'\n',roomda[:5])
print(roomhd.index('Size'))

floor=roomhd.index('Floor')
id_=roomhd.index('Id')
price=roomhd.index('Price')
size=roomhd.index('Size')
year=roomhd.index('Year')
district=roomhd.index('District')

for item in roomda:
    floor_lst.append(float(item[floor])) 
for item in roomda:
    id_lst.append(float(item[id_])) 
for item in roomda:
    price_lst.append(float(item[price])) 
for item in roomda:
    size_lst.append(float(item[size])) 
for item in roomda:
    year_lst.append(float(item[year])) 
for item in roomda:
    district_lst.append(item[district])

stata_head=['Floor','Id','Price','Size','Year']
all_da=[floor_lst,id_lst,price_lst,size_lst,year_lst]
stata_count=[]
stata_max=[]
stata_min=[]
stata_mean=[]
stata_std=[]
stata_25=[]
stata_50=[]
stata_75=[]
stata_majority=[]
stata_skew=[]
stata_kurt=[]

#求标准差  
def std(item):  
    s1=0  
    s2=0
    for i in item:
        s1+=(i-float(sum(item))/len(item))**2;  
        s2+=1
    return (float(s1)/float(s2))**0.5
#求中位数  
def s_50(item):  
    item=sorted(item);  
    if len(item)%2==1:  
        return item[int(len(item)/2)]
    else:  
        return (item[int(len(item)/2-1)]+item[int(len(item)/2)])/2  
#已知列表长度为23677求下四分位数
def s_25(item):  
    item=sorted(item) 
    return item[int(len(item)/4-1)]*0.75+item[int(len(item)/4)]*0.25  
#已知列表长度为23677求上四分位数
def s_75(item):  
    item=sorted(item) 
    return item[int(len(item)/4*3-1)]*0.25+item[int(len(item)/4*3)]*0.75
#众数
def majorityElement(item):
    times=0
    nums=0
    for i in item:
        if item.count(i)>times:
            nums=i
            times=item.count(i)
    return nums
#求偏度  
def skew(item):  
    s1=0  
    s2=0
    s3=0
    for i in item:
        s1+=(i-float(sum(item))/len(item))**3  
        s2+=1
        s3+=(i-float(sum(item))/len(item))**2
    return (float(s1)/float(s2))/((float(s3)/float(s2))**1.5)
#求峰度  
def kurt(item):  
    s1=0  
    s2=0
    s3=0
    for i in item:
        s1+=(i-float(sum(item))/len(item))**4  
        s2+=1
        s3+=(i-float(sum(item))/len(item))**2
    return (float(s1)/float(s2))/((float(s3)/float(s2))**2)-3

for item in all_da:
    stata_count.append(len(item))
    stata_max.append(max(item))
    stata_min.append(min(item))
    stata_mean.append(float(sum(item))/len(item))
    stata_std.append(std(item))
    stata_25.append(s_25(item))
    stata_50.append(s_50(item))
    stata_75.append(s_75(item))
    stata_majority.append(majorityElement(item))
    stata_skew.append(skew(item))
    stata_kurt.append(kurt(item))
```
