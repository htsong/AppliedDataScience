```python
def read_csv(file_name):
    lianjia_data=[]
    with open(file_name,encoding='utf-8') as csvf:
        lianjia_data=[line.strip().split(',') for line in csvf.readlines() if line.strip() !='']
    head,data=lianjia_data[0],lianjia_data[1:]
    return(head,data)
file_name="lianjia.csv"
hd,dt=read_csv(file_name)
```

```python
#求size和price 的最大值和最小值及均值
size_idx=hd.index('Size')
price_idx=hd.index('Price')
size_lst=[float(i[size_idx]) for i in dt]#if float(i[size_idx])>9]
price_lst=[float(j[price_idx])for j in dt]#if float(j[price_idx])>60]
mean_price=sum(price_lst)/len(price_lst)
mean_size=sum(size_lst)/len(size_lst)
print("The max of price is {},The min of price is {}\nThe max of size is {},The min of size is {}".format(max(price_lst),min(price_lst),max(size_lst),min(size_lst)))
print('the average of price is {}, the average of size is {}'.format(mean_price,mean_size))
```

```python
#价格和面积的单元回归分析
def Parameter_SLR(x,y):
    dinominator=0
    numerator=0
    for i in range(len(x)):
        numerator = numerator+(x[i] - sum(x)/len(x))*(y[i] - sum(x)/len(x))
        dinominator = dinominator+(x[i] - sum(x)/len(x))**2
        b1 = float(numerator)/float(dinominator)
        b0 =float(sum(y)/len(y))-float(sum(x)/len(x))*b1
        return b0, b1
x,y=size_lst,price_lst
b0,b1=Parameter_SLR(x,y)
print('linear regression with price and size: price={}*size+{}'.format(b1,b0))
#系数为负，表明房子价格和面积成反比。违背常理，说明还有其他变量影响价格，但是我不知道在不导入高级库的情况下如何求多元回归的系数。
```
