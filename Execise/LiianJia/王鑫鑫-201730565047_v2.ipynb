```python
#渣渣数据可视化
fname=input("Enter your file:")  #输入文件：lianjia.csv
all_list=[]

with open(fname,encoding="utf-8") as csvfile:
    for line in csvfile.readlines():
        if line.strip() != "":
            all_list.append(line.strip().split(","))
            
without_head=all_list[1:]
year=list()
for line in without_head:
    year.append(line[11])
year1=sorted(year)     #必须排序

def statas(numlist):   #统计频数
    count=0
    i=1
    frequence=[]
    while i <len(numlist):
        if numlist[i]==numlist[i-1]:
            count=count+1
        else:
            frequence.append(count+1)
            count=0
        i=i+1
    frequence.append(count+1)
    return frequence 
fre_year=statas(year1)

def list_what(numlist):
    a=list()
    a.append(numlist[0])
    i=1
    while i < len(numlist):
        if numlist[i]!=numlist[i-1]:
            a.append(numlist[i])
        i=i+1
    return a
list_year=list_what(year1)

po_year=[]
for i in fre_year:
    i1=int(int(i)/13)   #一个点表示13次
    po_year.append(i1)


k=0
for y in list_year:
    j=0
    print(y+":",end="")
    while j < int(po_year[k]):
        print("*",end="")
        j=j+1
    print()
    k=k+1
```
