```python
lianjia_flnm='lianjia.csv'
def py_read_csv(file_name):
    room_data=[]
    with open(file_name,encoding='utf-8') as csvfile:
        for line in csvfile.readlines():
            if line.strip() != '':
                room_data.append(line.strip().split(','))

    room_header=room_data[0]
    room_data=room_data[1:]
    return(room_header,room_data)
    csvfile.close()
#描述性统计分析
def title_list(title_name):#求某类型特征数值列表
    hd,dt=py_read_csv(lianjia_flnm)
    idx=hd.index(title_name)
    lst=[]
    for item in dt:
        lst.append(float(item[idx]))
    return lst

def count(name_list):#求数值个数
    count=[]
    for i in name_list:
        count.append(len(i))
    return count

def su_m(name_list):#求数值总和
    su_m=[]
    for i in name_list:
        su_m.append(sum(i))
    return su_m

def mean(name_list):#求数值平均数
    mean=[]
    for i in name_list:
        mean.append(sum(i)/len(i))
    return mean

def var(name_list):#求数值方差
    var=[]
    for i in name_list:
        b=0
        for j in i:
            a=float((j-sum(i)/len(i))**2)
            b=a+b
        var.append(b/len(i))
    return var

def std(name_list):#求数值标准差
    std=[]
    for i in name_list:
        b=0
        for j in i:
            a=float((j-sum(i)/len(i))**2)
            b=a+b
        std.append((b/len(i))**(1/2))
    return std
    
def sk(name_list):#求数值偏度
    sk=[]
    for i in name_list:
        b=0
        e=0
        for j in i:
            a=float((j-sum(i)/len(i))**2)
            w=float((j-sum(i)/len(i))**3)
            b=a+b
            e=e+w
        sk.append(e/len(i)/((b/len(i))**(3/2)))
    return sk

def bk(name_list):#求数值峰度
    bk=[]
    for i in name_list:
        o=0
        b=0
        for j in i:
            a=float((j-sum(i)/len(i))**2)
            k=float((j-sum(i)/len(i))**4)
            b=a+b
            o=o+k
        bk.append(o/len(i)/((b/len(i))**2)-3)
    return bk

def mi_n(name_list):#求最小值
    mi_n=[]
    for i in name_list:
        mi_n.append(min(i))
    return mi_n

def ma_x(name_list):#求最大值
    ma_x=[]
    for i in name_list:
        ma_x.append(max(i))
    return ma_x

def mode(name_list):#求众数
    mode=[]
    for i in name_list:
        counts=dict()
        for j in i:
            counts[j]=counts.get(j,0)+1
        bigcount=None
        bigword=None
        for word,coun_t in counts.items():
            if bigcount is None or coun_t>bigcount:
                bigword=word
                bigcount=coun_t
        mode.append(bigword)
    return mode

def Q1(name_list):#求下四分位数
    a25=[]
    for i in name_list:
        i.sort()
        Q1=float((len(i)+1)/4)
        if len(i)%2!=0:
            a25.append(i[int(Q1-1)])
        else:
            a25.append((Q1-int(Q1))*i[int(Q1)]+(int(Q1)+1-Q1)*i[int(Q1)-1])
    return a25

def Q2(name_list):#求中位数
    a50=[]
    for i in name_list:
        i.sort()
        Q2=float((len(i)+1)/2)
        if len(i)%2!=0:
            a50.append(i[int(Q2-1)])
        else:
            a50.append((i[len(i)/2-1]+i[len(i)/2])/2)
    return a50

def Q3(name_list):#求上四分位数
    a75=[]
    for i in name_list:
        i.sort()
        Q3=float((len(i)+1)*3/4)
        if len(i)%2!=0:
            a75.append(i[int(Q3-1)])
        else:
            a75.append((Q3-int(Q3))*i[len(i)-1]+(int(i)+1-Q3)*i[int(Q3)-1])
    return a75

Floor_list=title_list('Floor')
Id_list=title_list('Id')
Price_list=title_list('Price')
Size_list=title_list('Size')
Year_list=title_list('Year')
lis_t=[Floor_list,Id_list,Price_list,Size_list,Year_list]
print('type:  ',' Floor ',' Id ',' Price ',' Size ',' Year ')
print('count:',count(lis_t))
print('sum:  ',su_m(lis_t))
print('mean: ',mean(lis_t))
print('mode: ',mode(lis_t))
print('var:  ',var(lis_t))
print('std:  ',std(lis_t))
print('max:  ',ma_x(lis_t))
print('min:  ',mi_n(lis_t))
print('25%:  ',Q1(lis_t))
print('50%:  ',Q2(lis_t))
print('75%:  ',Q3(lis_t))
print('bk:   ',bk(lis_t))
print('sk:   ',sk(lis_t))
```
