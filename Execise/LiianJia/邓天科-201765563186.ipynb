```python
lianjia_flnm='lianjia.csv'
def py_read_csv(file_name):
    room_data=[]
    with open(file_name,encoding='utf-8') as csvfile:
        for line in csvfile.readlines():
            if line.strip() != '':
                room_data.append(line.strip().split(','))

    room_header=room_data[0]
    room_data=room_data[1:]
    return(room_header,room_data)
    csvfile.close()
#描述性统计分析
hd,dt=py_read_csv(lianjia_flnm)
s=['Floor','Id','Price','Size','Year']
count=[]
su_m=[]
mean=[]
var=[]
std=[]
mi_n=[]
a25=[]
a50=[]
a75=[]
ma_x=[]
mode=[]
bk=[]
sk=[]
for i in s:
    idx=hd.index(i)
    lst=[]
    for item in dt:
        lst.append(float(item[idx]))
    count.append(len(lst))
    su_m.append(sum(lst))
    mean.append(sum(lst)/len(lst))
    o=0
    b=0
    e=0
    for j in lst:
        a=float((j-sum(lst)/len(lst))**2)
        k=float((j-sum(lst)/len(lst))**4)
        w=float((j-sum(lst)/len(lst))**3)
        b=a+b
        o=o+k
        e=e+w
    var.append(b/len(lst))
    std.append((b/len(lst))**(1/2))
    bk.append(o/len(lst)/((b/len(lst))**2)-3)
    sk.append(e/len(lst)/((b/len(lst))**(3/2)))
    mi_n.append(min(lst))
    lst.sort()
    Q1=float((len(lst)+1)/4)
    Q2=float((len(lst)+1)/2)
    Q3=float((len(lst)+1)*3/4)
    if len(lst)%2!=0:
        a25.append(lst[int(Q1-1)])
        a50.append(lst[int(Q2-1)])
        a75.append(lst[int(Q3-1)])
    else:
        a25.append((Q1-int(Q1))*lst[int(Q1)]+(int(Q1)+1-Q1)*lst[int(Q1)-1])
        a50.append((lst[len(lst)/2-1]+lst[len(lst)/2])/2)
        a75.append((Q3-int(Q3))*lst[len(lst)-1]+(int(lst)+1-Q3)*lst[int(Q3)-1])
    ma_x.append(max(lst))
    counts=dict()
    for v in lst:
        counts[v]=counts.get(v,0)+1
    bigcount=None
    bigword=None
    for word,coun_t in counts.items():
        if bigcount is None or coun_t>bigcount:
            bigword=word
            bigcount=coun_t
    mode.append(bigword)
    lst.clear()
print('type:',s)
print('count:',count)
print('sum: ',su_m)
print('mean:',mean)
print('var: ',var)
print('std: ',std)
print('min: ',mi_n)
print('25%: ',a25)
print('50%: ',a50)
print('75%: ',a75)
print('max: ',ma_x)
print('mode:',mode)
print('bk:  ',bk)
print('sk:  ',sk)
```

```python
#单价最低的二手房有哪些？
price=[]
size=[]
perprice=[]
name=[]
hd,dt=py_read_csv(lianjia_flnm)
idx=hd.index('Price')
edx=hd.index('Size')
for item in dt:
    price.append(float(item[idx]))
    size.append(float(item[edx]))
for j in list(range(len(price))):
    perprice.append(price[j]/size[j])
mi_n=min(perprice)
print('最低单价是：',mi_n)
for j in list(range(len(perprice))):
    if perprice[j]==mi_n:
        name.append(dt[j])
print(name)
```
