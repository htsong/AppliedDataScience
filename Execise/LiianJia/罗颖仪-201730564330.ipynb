```python
lianjia_flnm="lianjia.csv"
def py_read_csv(file_name):
    room_data=[]
    with open(file_name,encoding="utf-8") as csvfile:
        #room_data=[line.strip().split(",")for line in csvfile.readlines( )if line.strip()!=""]
        for line in csvfile.readlines():
            if line.strip()!="":room_data.append(line.strip().split(","))
    room_header,room_data=room_data[0],room_data[1:]
    return(room_header,room_data)
hd,dt=py_read_csv(lianjia_flnm)
print(hd)
print("\n",dt[:5])
```

```python
#求众数
def mode(lst):
    count_dict={}
    for i in lst:
        if i in count_dict:
            count_dict[i]+=1
        else:
            count_dict[i]=1
    max_appear=0
    for v in count_dict.values():
        if v>max_appear:
            max_appear=v
    if max_appear==1:
        return
    mode_list=[]
    for k,v in count_dict.items():
        if v==max_appear:
            mode_list.append(k)
    return mode_list     

#求四分位数
def median(x):
    x=sorted(x)
    length=len(x)
    mid,rem=divmod(length,2)#divmod函数返回商和余数
    if rem:
        return x[:mid],x[mid+1:],x[mid]
    else:
        return x[:mid],x[mid:],(x[mid-1]+x[mid])/2


#求偏度
def skew(xi):
    n=len(xi)
    xhead=sum(xi)/n   #求均值
    xd=0
    for i in xi:
        xd+=(i-xhead)**2  #求方差
    xd_mean=xd/n   #求均方差
    x3=0
    for j in xi:
        x3+=(j-xhead)**3
    u3=x3/n  #求3阶中心距
    g1=(n**2*u3)/((n-1)*(n-2)*xd_mean**3)  #求偏度
    return(g1)

#求峰度
def kurt(xi):
    n=len(xi)
    xhead=sum(xi)/n   #求均值
    xd=0
    for i in xi:
        xd+=(i-xhead)**2  #求方差
    xd_mean=xd/n   #求均方差
    x4=0
    for j in xi:
        x4+=(j-xhead)**4
    u4=x4/n  #求4阶中心距
    g2=(n**2*(n+1)*u4)/((n-1)*(n-2)*(n-3)*xd_mean**4)-((3*(n-1)**2)/((n-2)*(n-3)))  #峰度
    return(g2)

idx=hd.index("Size")
size_lst=[float(item[idx])for item in dt]

print("【面积】")
print("最大值:{},最小值:{},平均值:{}".format(max(size_lst),min(size_lst),sum(size_lst)/len(size_lst)))
print("众数:{}".format(mode(size_lst)))
lHalf,rHalf,q2=median(size_lst)
print("前25%:{}，前50%:{}，前75%:{}".format(median(lHalf)[2],q2,median(rHalf)[2]))
print("偏度:{}".format(skew(size_lst)))
print("峰度:{}".format(kurt(size_lst)))

idx=hd.index("Floor")
floor_lst=[float(item[idx])for item in dt]

print("【层】")
print("最大值:{},最小值:{},平均值:{}".format(max(floor_lst),min(floor_lst),sum(floor_lst)/len(floor_lst)))
print("众数:{}".format(mode(floor_lst)))
lHalf,rHalf,q2=median(floor_lst)
print("前25%:{}，前50%:{}，前75%:{}".format(median(lHalf)[2],q2,median(rHalf)[2]))
print("偏度:{}".format(skew(floor_lst)))
print("峰度:{}".format(kurt(floor_lst)))

idx=hd.index("Price")
price_lst=[float(item[idx])for item in dt]

print("【价格】")
print("最大值:{},最小值:{},平均值:{}".format(max(price_lst),min(price_lst),sum(price_lst)/len(price_lst)))
print("众数:{}".format(mode(price_lst)))
lHalf,rHalf,q2=median(price_lst)
print("前25%:{}，前50%:{}，前75%:{}".format(median(lHalf)[2],q2,median(rHalf)[2]))
print("偏度:{}".format(skew(price_lst)))
print("峰度:{}".format(kurt(price_lst)))

idx=hd.index("Year")
year_lst=[float(item[idx])for item in dt]

print("【年份】")
print("最大值:{},最小值:{},平均值:{}".format(max(year_lst),min(year_lst),sum(year_lst)/len(year_lst)))
print("众数:{}".format(mode(year_lst)))
lHalf,rHalf,q2=median(year_lst)
print("前25%:{}，前50%:{}，前75%:{}".format(median(lHalf)[2],q2,median(rHalf)[2]))
print("偏度:{}".format(skew(year_lst)))
print("峰度:{}".format(kurt(year_lst)))

idx=hd.index("Id")
id_lst=[float(item[idx])for item in dt]

print("【ID】")
print("最大值:{},最小值:{},平均值:{}".format(max(id_lst),min(id_lst),sum(id_lst)/len(id_lst)))
print("众数:{}".format(mode(id_lst)))
lHalf,rHalf,q2=median(id_lst)
print("前25%:{}，前50%:{}，前75%:{}".format(median(lHalf)[2],q2,median(rHalf)[2]))
print("偏度:{}".format(skew(id_lst)))
print("峰度:{}".format(kurt(id_lst)))
```
