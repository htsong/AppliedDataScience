```python
def descriptive_analysis(name,target,fil): #描述性分析
    Available=["Floor","Price","Size","Year","Id"]
    
    if name not in Available:
        return "写错了！！！！"
    name_index = fil[0].index(name)
    name_list = [float(c[name_index]) for c in fil[1:]]
    name_max = max(name_list)
    name_min = min(name_list)
    name_mean = sum(name_list)/len(name_list)
    name_observation = len(name_list)
    name_variance = sum([(c-name_mean)**2 for c in name_list])/len(name_list)
    name_std = (sum([(c-name_mean)**2 for c in name_list])/len(name_list))**(1/2)
    name_count = len(name_list)
    name_quartile = [(sorted(name_list)[int(name_count*i/4)]+sorted(name_list)[int(name_count*i/4+1)])/2 if name_count %2 == 0 else sorted(name_list)[int(name_count*i/4)] for i in range(1,4)]
    name_skewness = sum([(c-name_mean)**3 for c in name_list])/(name_count*name_std**3)
    name_kurtosis = sum([(c-name_mean)**4 for c in name_list])/(name_count*name_std**4)
    number = max([name_list.count(b) for b in name_list])
    a = []
    for i in name_list:
        if name_list.count(i) == number :
            a.append(i)
        
    name_mode = list(set(a))

    
    #字典
    name_available = {"name_list":name_list,
    "name_max":name_max,
    "name_min":name_min,
    "name_mean":name_mean,
    "name_observation":name_observation,
    "name_variance":name_variance,
    "name_std":name_std,
    "name_count":name_count,
    "name_quartile":name_quartile,
    "name_skewness":name_skewness,
    "name_kurtosis":name_kurtosis,
    "name_mode":name_mode
    }

    for c in name_available:
        if c == "name"+"_"+target:
            return name_available[c]
```

```python
def year(fil):#弄出Year#目标：运用本处的坐标，得到一个直方图,观察一下不同年代房子的情况（由于1950-1990的房子年代过于久远，所以放在一起观察）、
    inx = fil[0].index("Year")
    years = [float(i[inx]) for i in fil[1:]]
    interval ={"1950 to 1990":0,
            "1991-2000":0,
            "2001-2010":0,
            "2011-2020":0}#确定字典中不同年份数量
    for i in years:   # 查找csv文件中不同层次房子的数量
        for b in interval:
            if i >= float(b[:4]) and i <= float(b[-4:]):
                interval[b]+=1
    yea = list(interval.values())#确定图像的最后一个行输出对象（x坐标）
    #将图像的x轴坐标做好，
    picture = []
    while len(picture)<121:
        picture.append([])
    picture.append([])
    picture[-1].append(" ".center(11))
    for b in interval:
        picture[-1].append(b.center(16))
    picture[-2].append("0".center(10))
    picture[-2].append("!")
    picture[-2].append("-"*64)
    i = 0
    e = ""
    while i <120:    #确定y坐标轴以及y上的图像
        picture[i].append(str((120-i)*100).center(10))
        picture[i].append("!")
        if (120-i) < int(yea[0]/100):
            picture[i].append("d"*16)
        else:
            picture[i].append("-"*16)
        if (120-i) < int(yea[1]/100):
            picture[i].append("a"*16)
        else:
            picture[i].append("-"*16)
        if (120-i) < int(yea[2]/100):
            picture[i].append("b"*16)
        else:
            picture[i].append("-"*16)
        if (120-i) < int(yea[3]/100):
            picture[i].append("c"*16)
        else:
            picture[i].append("-"*16)
        i+=1
    b = 0
    for i in picture:#汇聚成字符串输出
        while b <= len(i)-1:
            e = e + i[b]
            b+=1
        e = e +"\n"
        b = 0
    return e
```
