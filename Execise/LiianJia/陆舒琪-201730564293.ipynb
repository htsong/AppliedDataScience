```python
filename = 'lianjia.csv'
def read_csv(file_name):
    room_data = []
    with open(file_name, encoding = 'utf-8') as csvfile:
        for line in csvfile.readlines():
            if line.strip() != '': room_data.append(line.strip().split(','))
    room_header, room_data = room_data[0], room_data[1:]
    return room_header, room_data
hd, dt = read_csv(filename)
lst = ['Floor','Id','Price','Size','Year']
def add(title):
    lst1 = []
    for i in dt: lst1.append(float(i[hd.index(title)]))
    return lst1 #返回每个指标下数据组成的列表
def count(title): return len(add(title))
def mean(title): return round(sum(add(title))/count(title),6)
def var(title):
    n = 0; a = mean(title); b = count(title)
    for i in add(title): n += (i- a)**2
    return round((n/b),6)
def max1(title): return max(add(title))
def min1(title): return min(add(title))
def quart(title):
    if count(title)%2 == 1:
        n = sorted(add(title))[int((count(title)+1)/4-1)]
    else:
        n = sorted(add(title))[int(count(title)/4-1)]
    return n
def mode(title):
    d = {}
    for i in add(title): d[i] = d.get(i,0)+1
    return list(d.keys())[list(d.values()).index(max(list(d.values())))]
def skewness(title):
    n = 0; a = mean(title)
    for i in add(title): n += (i - a)**3
    return round(n/var(title)**1.5/count(title),6)
def kurtosis(title):
    n = 0; a = mean(title)
    for i in add(title): n += (i - a)**4
    return round(n/var(title)**2/count(title),6)
def call_func(func_name, title_list):
    lst2 = []
    for i in title_list: lst2.append(func_name(i))
    return lst2
print(f'''均值{call_func(mean,lst)}
方差{call_func(var,lst)}
最大值{call_func(max1,lst)}
最小值{call_func(min1,lst)}
四分位数{call_func(quart,lst)}
众数{call_func(mode,lst)}
偏度{call_func(skewness,lst)}
峰度{call_func(kurtosis,lst)}''')

'''统计不同户型和装修情况的平均单价
 思路：先求出每套房的单价
       将装修、户型、单价三个列表压缩成一个新列表
       将重新排序后的列表中的装修和户型信息用字符串连接，合并成一个新的列表，创建字典对其计数；创建另一个列表，包含对应的单价
       通过字典的键和值计算不同户型和装修的二手房的平均单价并呈现'''
def add1(title):
    lst1 = []
    for i in dt: lst1.append(i[hd.index(title)])
    return lst1 
perprice = []; plst = add('Price'); slst = add('Size'); rlst = add1('Renovation'); llst = add1('Layout'); n = 0
for i in plst: 
    perprice.append(i/slst[n])#每套房单价 
    n += 1
lst1 = [];  n = 0
lst2 = sorted(list(zip(rlst,llst,perprice)))
lst3 = []; lst4 = []
for i in lst2:
    lst3.append(f'{i[0]},{i[1]}')
    lst4.append(float(i[2]))
d = {}; lst5 = []; n = 0
for i in lst3: d[i] = d.get(i,0) + 1
for k,v in d.items():
    lst5.append([sum(lst4[n:n+v])/v,k])
    n += v
for i in sorted(lst5):
    print('*'*int(i[0]*8),i[1],' 单价',round(i[0],2))
```
