```python
fhand=open("lianjia.csv",encoding='utf-8')
L=[]
for line in fhand.readlines():
    if line is not '\n':
        L.append(line.strip().split(','))
S=[]
P=[]
Y=[]
num1,sum1,sum2,sum3=0,0,0,0
s=L[0].index('Size')
p=L[0].index('Price')
y=L[0].index('Year')
for i in range(1,len(L)):
    s1=L[i]
    S.append(float(s1[s]))
    num1=num1+1    
for j in range(1,len(L)):
    p1=L[j]
    P.append(float(p1[p]))    
for k in range(1,len(L)):
    y1=L[k]
    Y.append(float(y1[y]))
#均值
def average(x):
    sum=0
    for i in range(len(x)):
        sum=sum+x[i]
    return sum/num1
#方差
def var(x):
    v=0
    for i in range(len(x)):
        q=(x[i]-average(x))**2
        v=v+q
    return v/(num1-1)
#众数
def zhongshu(x):
    count=dict()
    for i in x:
        count[i]=count.get(i,0)+1
    return max(count.values())
#四分位数
def sifenwei(x):
    sfw=[]
    x.sort()
    if len(x)%2 == 0:
        weizhi1=int((len(x)+1)*0.25)
        weizhi2=int((len(x)+1)*0.5)
        weizhi3=int((len(x)+1)*0.75)
        q1=0.25*x[weizhi1]+0.75*x[weizhi1+1]
        q2=(x[weizhi2]+x[weizhi2+1])/2
        q3=q1=0.25*x[weizhi3+1]+0.75*x[weizhi3]        
    else:
        weizhi1=int((len(x)+1)*0.25)     
        weizhi2=(len(x)+1)*0.5
        if weizhi2%2!=0:
            weizhi2=int(weizhi2)
            q2=(x[weizhi2]+x[weizhi2+1])/2
        else:q2=x[weizhi2]
        weizhi3=(len(x)+1)*0.75
        if weizhi3%2!=0:
            weizhi3=int(weizhi3)
            q3=(x[weizhi3]+x[weizhi3+1])/2
        else:q3=x[weizhi3]
        q1=x[weizhi1]
    return q1,q2,q3
     
#峰度
def feng(x):
    fenzi=0
    fenmu=0
    for i in range(len(x)):
        he1=(x[i]-average(x))**4
        fenzi=fenzi+he1
    for i in range(len(x)):
        he2=(x[i]-average(x))**2
        fenmu=fenmu+he2
    return ((fenzi/num1)/((fenmu/num1)**2))-3
#偏度
def pian(x):
    fenzi=0
    fenmu=0
    for i in range(len(x)):
        he1=(float(x[i])-average(x))**3
        fenzi=fenzi+he1
    for i in range(len(x)):
        he2=(float(x[i])-average(x))**2
        fenmu=fenmu+he2
    return ((fenzi/num1)/((fenmu/(num1-1))**(3/2)))
print('面积,房价,年份的均值分别为%f,%f,%f'% (average(S),average(P),average(Y)))
print('面积,房价,年份的方差分别为%f,%f,%f'% (var(S),var(P),var(Y)))
print('面积,房价,年份的最大值分别为%f,%f,%f'% (max(S),max(P),max(Y)))
print('面积,房价,年份的最小值分别为%f,%f,%f'% (min(S),min(P),min(Y)))
print('面积,房价,年份的众数分别为%f,%f,%f'% (zhongshu(S),zhongshu(P),zhongshu(Y)))
print('面积,房价,年份的四分位分别为%s、%s、%s'% (sifenwei(S),sifenwei(P),sifenwei(Y)))
print('面积,房价,年份的峰度分别为%f,%f,%f'% (feng(S),feng(P),feng(Y)))
print('面积,房价,年份的偏度分别为%f,%f,%f'% (feng(S),feng(P),feng(Y)))


#问题：住房的房价与年份间的相关关系是怎样的？假设年份为自变量。
#求β1
def xishu1(x,y):
    fenzi=0
    fenmu=0
    for i in range(len(x)):
        q=(x[i]-average(x))*(y[i]-average(y))
        fenzi=fenzi+q
        p=(x[i]-average(x))**2
        fenmu=fenmu+p   
    return fenzi/fenmu
#求β0
def xishu2(x,y):
    return average(y)-xishu1(x,y)*average(x)
#求R方
def rsquare(x,y):
    sst=0
    sse=0
    guji=[]
    for i in range(len(x)):
        q=(y[i]-average(y))**2
        sst=sst+q
    for i in range(len(x)):
        guji.append(x[i]*xishu1+xishu2)
    for i in range(len(guji)):
        p=(guji[i]-average(y))**2
        sse=sse+p
    return sse/sst
print('房价与年份的关系为房价=%f年份+%f,他们之间的R方为%f'% (xishu1(Y,P),xishu2(Y,P),rsquare(Y,P))
        
```
