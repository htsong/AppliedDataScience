```python
#显示表格数据
import pandas as pd
df=pd.read_csv("lianjia.csv")
df
```

```python
#输出表头
file=open("lianjia.csv",encoding="utf-8")
roomdata=[]
for line in file.readlines():
    if line.strip()!="":
        roomdata.append(line.strip().split(","))
hd,dt=roomdata[0],roomdata[1:]
print(hd)
file.close()
```

```python
#求全部房屋价格的均值…
price=[]
sum=0
for row in dt:
    price.append(float(row[-5]))
    sum=sum+float(row[-5])
num=len(price)
avgprice=sum/num
print("price mean:","%.2f"%avgprice)
```

```python
#房屋价格的标准差与方差
price_std=0
variance=0
for data in price:
    variance=(float(data)-avgprice)**2+ variance
price_variance=variance/len(dt)
price_std=(price_variance)**(1/2)
print("price variance:","%.2f"%variance)
print("princ standard deviation:","%.2f"%price_std)
```

```python
#房屋价格的中位数
def get_median(price):
    price=sorted(price)
    leng=len(price)
    if leng %2==0:
        median=(price[leng//2]+price[leng//2-1])/2
        price[0]=median
    if leng%2==1:
        median=price[(leng-1)//2]
        price[0]=median
        return price[0]
print("median price:",get_median(price))
```

```python
#求对应序号房子的单位价值（单位：元/平方米）
size=[]
for row in dt:
    size.append(float(row[-2]))
for i in range(len(dt)):
    unit_price=price[i]/size[i]
    funit_price=float("%.2f"%unit_price)
    print("ID",i+1,"uprice:",funit_price)
```

```python
#最高房价与对应的房子信息，最低价与对应房屋信息
maxprice=max(price)
maxprice_name=price.index(maxprice)
print("maxprice:",maxprice,"roomID:",maxprice_name)
minprice=min(price)
minprice_name=price.index(minprice)
print("minprice:",minprice,"roomID:",minprice_name)
```

```python
#房价的众数
price_set=set(price)
roomdict=dict()
for i in price_set:
    roomdict[i]=price.count(i)
price_mode=[]
for key,value in roomdict.items():
    if value==max(roomdict.values()):
        price_mode.append(key)
for num in price_mode:
    price_mode=num
    print("price mode:",price_mode)
```
